```python
sn_group_count = purchase_data.groupby(['SN',]).size().reset_index(name='count')
sn_group_count.shape
```

```python
sn_group = purchase_data.groupby("SN")

sn_purchase_df = pd.DataFrame(sn_group["Price"].sum())
sn_purchase_df.shape
```

```python
merge_df = pd.merge(sn_group_count, sn_purchase_df, on="SN")
merge_df
```

```python
merge_df["Average Purchase Price"] = merge_df["Price"] / merge_df["count"] 
merge_df
```

```python
sort_by_value = merge_df.sort_values('Price',
                                     ascending=False)
sort_by_value
```

```python
rename_df = sort_by_value.rename(columns={"Price": "Total Purchase Value"})
rename_df
```
